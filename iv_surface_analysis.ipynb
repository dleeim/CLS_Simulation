## 1. Load data

```python
df = pd.read_csv("data/modified/binance_option.csv")
df['date'] = pd.to_datetime(df['date'])

print("df head")
display(df.head())

print("df describe")
display(df.describe())

print("df info")
display(df.info())
```

## 2. Surface Volatility Inspired (SVI) and Spline Interpolation
* The code is inspired by **joshuapjacob/crypto‑volatility‑surface** `surface.py`

### 2.1 Methodology
* Pick a date: **`2025-05-18 00:00:00`**
* Fit parameters of SVI to option chain data 
* Smooth out the SVI model with Spline Interpolation using **`RectBivariateSpline`**

```python
# Choose a date
snap_mask   = df["date"] == pd.Timestamp("2023-05-19 00:00:00+00:00", tz="UTC")
df_snapshot = df[snap_mask].copy()

# Build the spline once from your data
spline = build_iv_surface_spline(df_snapshot)
```

## 3. Plot IV Surface

```python
# Then use it for any arbitrary grids
k_new = np.linspace(-0.8, 0.6, 20)  # arbitrary log-moneyness values
T_new = np.linspace(0.05, 0.3, 20)    # arbitrary time-to-expiry values

# Get interpolated surface (shape will be 15x8)
Z_interpolated = spline(k_new, T_new)

# 3-D IV Surface
plot_iv_surface(k_new,T_new,Z_interpolated)
```

## 4. Plot IV Skew

```python
# Then use it for any arbitrary grids
k_new = np.linspace(-0.8, 0.6, 20)  # arbitrary log-moneyness values
T_new = np.array([0.03, 0.15, 0.30])

# Get interpolated surface (shape will be 15x8)
Z_interpolated = spline(k_new, T_new)

# IV Skew
plot_iv_skew(k_new,T_new,Z_interpolated)
```

```python
# Then use it for any arbitrary grids
k_new = np.array([0.0])  # arbitrary log-moneyness values
T_new = np.linspace(0.05, 0.3, 20)    # arbitrary time-to-expiry values

# Get interpolated surface (shape will be 15x8)
Z_interpolated = spline(k_new, T_new)

# IV Skew
plot_term_structure(k_new,T_new,Z_interpolated)
```
